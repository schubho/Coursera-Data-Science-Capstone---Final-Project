```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
import numpy as np
import urllib.request
import urllib
```

```python
# Step 1 - obtaining a list of Mumbai neighbourhoods and their corresponding wiki links

fnm = "https://en.wikipedia.org/wiki/Category:Neighbourhoods_in_Mumbai"

f = urllib.request.urlopen(fnm)
myfile = f.read()
soup = BeautifulSoup(myfile, 'lxml')

print(soup.prettify)
```

```python
#creating a dictionary file with the neighborhood details
mum = dict()
for link in soup.find_all('a', href=True):
    if link['href'][0:6] == '/wiki/':
        print(link.text.strip(), 'https://en.wikipedia.org'+link['href'])
        mum[link.text.strip()] = 'https://en.wikipedia.org'+link['href']
```

<h4> Processing 138 Mumbai neighbourhoods - visiting their respective wiki links and obtaining corresponding latitude and longitide details</h4>

```python
import urllib
```

```python
count = 136
lldict7 = dict()

while count > -1:
    fnm = mum_hood.iloc[count,1]
    print(fnm)
    f = urllib.request.urlopen(fnm)
    myfile = f.read()
    soup7 = BeautifulSoup(myfile, 'lxml')
    llgeo = soup7.find("span", attrs={ "class" : "geo"})
    if llgeo == None:
        count = count - 1
        print(count)
        continue
    else:
        ll = llgeo.text.split('; ')
        hood_lat = ll[0]
        print(hood_lat)
        hood_long = ll[1]
        print(hood_long)
        lldict7[mum_hood.iloc[count,0]] = (hood_lat, hood_long)
        count = count - 1
        print(count)
    
```

```python
# creating a dictionary with all Mumbai neighbourhoods and their corresponding latitude and longitude details
lldict7
```

```python
ll_hood = pd.DataFrame.from_dict(lldict7, orient='index')
```

```python
ll_hood.reset_index(inplace=True)
```

```python
# create map of Mumbai using latitude and longitude values
map_mumbai = folium.Map(location=[19.0760, 72.8777], zoom_start=11)

# add markers to map
for lat, lng, neighborhood in zip(hood_df['latitude'], hood_df['longitude'], hood_df['neighborhood']):
    lat = float(lat)
    lng = float(lng)
    label = '{}'.format(neighborhood)
    label = folium.Popup(label, parse_html=True)
    folium.CircleMarker([lat, lng], radius=5, popup=label, color='blue', fill=True, fill_color='#3186cc', fill_opacity=0.7, parse_html=False).add_to(map_mumbai)
    
```
